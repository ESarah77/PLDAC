```python
import librosa
import matplotlib.pyplot as plt
```

```python
def pitch_shift(y, sr, shift=4):
    return librosa.effects.pitch_shift(y, sr=sr, n_steps=shift)
```

```python
def time_stretch(y, speed_rate=0.5):
    return librosa.effects.time_stretch(y, rate=speed_rate)
```

```python
y, sr = librosa.load("speech_commands_v0.02/bird/0a7c2a8d_nohash_0.wav")

fig, ax = plt.subplots(nrows=3, sharex=True)

librosa.display.waveshow(y, sr=sr, ax=ax[0], color="blue")
ax[0].set(title='Original')
ax[0].label_outer()

librosa.display.waveshow(pitch_shift(y, sr), sr=sr, ax=ax[1], color="blue")
ax[1].set(title='Pitch shift')
ax[1].label_outer()

librosa.display.waveshow(time_stretch(y), sr=sr, ax=ax[2], color="blue")
ax[2].set(title='Time stretch')
ax[2].label_outer()
```
